```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
gmf_hr = pd.read_csv('./data/gmf_HR.csv')
mlp_hr = pd.read_csv('./data/mlp_HR.csv')
mlp_pretrain_hr = pd.read_csv('./data/mlp_pretrain_HR.csv')

gmf_ndcg = pd.read_csv('./data/gmf_NDCG.csv')
mlp_ndcg = pd.read_csv('./data/mlp_NDCG.csv')
mlp_pretrain_ndcg = pd.read_csv('./data/mlp_pretrain_NDCG.csv')
```

```python
fig, ax = plt.subplots()
ax.plot(gmf_hr['Step'], gmf_hr['Value'], label='GMF-HR')
ax.plot(mlp_hr['Step'], mlp_hr['Value'], label='MLP-HR')
ax.plot(mlp_pretrain_hr['Step'], mlp_pretrain_hr['Value'], label='MLP-Pretrain-HR')

ax.set(xlabel='Epoch', ylabel='Hit Ratio@10',
       title='Hit Ratio: MLP performs slightly better than GMF after pretraining')
ax.grid()
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(gmf_ndcg['Step'], gmf_ndcg['Value'], label='GMF-HR')
ax.plot(mlp_ndcg['Step'], mlp_ndcg['Value'], label='MLP-HR')
ax.plot(mlp_pretrain_ndcg['Step'], mlp_pretrain_ndcg['Value'], label='MLP-Pretrain-HR')

ax.set(xlabel='Epoch', ylabel='NDCG@10',
       title='NDCG: MLP performs slightly better than GMF after pretraining')
ax.grid()
plt.legend()
plt.show()
```
